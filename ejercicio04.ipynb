```haskell
celsiusAFahrenheit :: Double -> Double
celsiusAFahrenheit c = c * 1.8 + 32

procesarLecturas :: (Double -> Double) -> [Double] -> [Double]
procesarLecturas f lista = map f lista
```

```haskell
procesarLecturas celsiusAFahrenheit [5, 10]
```

```haskell
-- Filtro personalizado
```

```haskell
crearFiltro :: (a -> Bool) -> ([a] -> [a])
crearFiltro condicion = \lista -> filter condicion lista

soloPares = crearFiltro even
soloPares [1..10] 
```

```haskell
-- función que genera “verificadores” de rangos.
```

```haskell
enRango :: Int -> Int -> (Int -> Bool)
enRango min max = \x -> x >= min && x <= max
entre5y10 = enRango 5 10
entre5y10 8
```

```haskell
-- funcion que eleva a una potencia dada
```

```haskell
potenciador :: Int -> (Int -> Int)
potenciador n = \base -> base ^ n

-- Ejemplo de uso:
-- alCuadrado = potenciador 2
-- alCubo = potenciador 3
-- alCuadrado 5 -> 25
-- alCubo 2     -> 8
```

```haskell
-- Sumar 5 a impares
```

```haskell
ajustarImpares :: [Int] -> [Int]
ajustarImpares = map (\x -> if odd x then x + 5 else x)
```

```haskell
ajustarImpares [1, 2, 3, 4, 5]
```

```haskell
-- Devuelve nombres mayores a 5 letras
```

```haskell
nombresLargos :: [String] -> [String]
nombresLargos = filter (\n -> length n > 5)
```

```haskell
nombresLargos ["Jose", "maria", "leonardo", "samuel", "jeronimo"]
```

```haskell
-- Sumar IVA a n >= 1000
```

```haskell
agregarIVA :: [Double] -> [Double]
agregarIVA xs = map (* 1.19) (filter (>= 1000) xs)
```

```haskell
agregarIVA [1200, 1000, 800, 600, 1250]
```

```haskell
-- Palabras que inician con vocal y su longitud
```

```haskell
longitudesVocales :: [String] -> [Int]
longitudesVocales ws = map length (filter (\w -> head w `elem` "aeiouAEIOU") ws)
```

```haskell
longitudesVocales ["area", "balon", "mesa", "aura"]
```

```haskell
-- cuadrados de los pares
```

```haskell
cuadradosPares :: [Int] -> [Int]
cuadradosPares xs = map (^2) (filter even xs)
```

```haskell
cuadradosPares [2, 3, 4, 5, 6]
```

```haskell
-- 4.
```

```haskell
-- Hay negativos en lista?
```

```haskell
hayNegativo :: [Int] -> Bool
hayNegativo = any (< 0)
```

```haskell
hayNegativo [-1, 2, 3, 5]
```

```haskell
-- todos los numeros son pares?
```

```haskell
todosPares :: [Int] -> Bool
todosPares = all even
```

```haskell
todosPares [1, 2, 3]
```

```haskell
-- nombres empiezan en mayuscula?
```

```haskell
import Data.Char (isUpper)
nombresCorrectos :: [String] -> Bool
nombresCorrectos = all (\n -> not (null n) && isUpper (head n))
```

```haskell
nombresCorrectos ["Luisa", "Jero", "Luis"]
```

```haskell
-- 5.
```

```haskell
-- elementos menores que 100.
```

```haskell
menoresQue100 :: [Int] -> [Int]
menoresQue100 = takeWhile (< 100)
```

```haskell
menoresQue100 [50, 60, 70, 80, 90, 98, 101, 102, 103]
```

```haskell
-- Quita los elementos 0 iniciales
```

```haskell
sinCerosIniciales :: [Int] -> [Int]
sinCerosIniciales = dropWhile (== 0)
```

```haskell
sinCerosIniciales [0, 0, 0, 0, 5, 0, 4, 6, 9]
```
